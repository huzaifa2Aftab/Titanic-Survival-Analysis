```python
df.drop( ["PassengerId"], axis=1, inplace=True)
```

```python
df
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Age"] = df["Age"].astype("int")
```

```python
sex = df["Sex"].value_counts()
fig, ax = plt.subplots()
ax.bar(sex.index, sex.values, color=["blue", "brown"], width=0.05)
plt.title("Gender Distribution")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

**577 Male and 314 Female**

```python
plt.hist(df['Age'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

*The age distribution analysis reveals that the most common ages among passengers were 29, 24, 22, 28, and 30*

```python
survive = df["Survived"].value_counts()
df_rename = survive.rename(index={0: 'Dead', 1: 'Survived'})
```

```python

fig, ax = plt.subplots()
ax.bar(df_rename.index, df_rename.values, color=["red", "green"],edgecolor='black', width=0.05)
plt.title("Survived Vs Dead Distribution")
plt.xlabel('Survival Status')
plt.ylabel("Count")
plt.show()
```

**549 Dead and 342 Survived**

```python
grouped = df.groupby(['Survived', 'Sex']).size()
grouped.rename(index={0: 'Dead', 1: 'Survived'})

```

```python
grouped = df.groupby(['Survived', 'Sex']).size()
gender_s = grouped.unstack()
gender_s.plot(kind='bar', stacked=True, width=0.05)
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.title('Survival Count by Gender')
plt.xticks(rotation=0)  # Rotate x-axis labels to horizontal
plt.xticks(ticks=[0, 1], labels=['Dead', 'Survived'], rotation=0)
plt.legend(title='Sex')
```
